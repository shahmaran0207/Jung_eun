```python
#csv파일 json파일 upload 및 encoding
def main():

    csvFile = "C:/Users/wjaud/Desktop/빅데이터 분석 경진대회/새 폴더/2011.csv"
    SiDodf = pd.read_csv("C:/Users/wjaud/Desktop/빅데이터 분석 경진대회/새 폴더/2011.csv", encoding = 'euc-kr')
    
    SiDo_geo = "C:/Users/USERP/Downloads/Si_Do_Ele_students_Teacher_ratio (1).csv"
```

```python
# folium.Map()을 활용하여 지도 설정
m = folium.Map(location=[36.45, 127.42],
                   tiles="OpenStreetMap", zoom_start=8)

# Choropleth 레이어를 만들고, 맵 m에 추가
    folium.Choropleth(
        geo_data = SiDo_geo,
        name='2011 범죄율 대비 신고율',
        data=SiDodf,
        columns=['행정구역별', '범죄율 대비 신고율'],
        key_on='feature.properties.CTP_KOR_NM', fill_color='PuRd',
        fill_opacity=0.7,
        line_opacity=0.2,
        legend_name='2011 범죄 대비 신고율').add_to(m)

    folium.LayerControl().add_to(m)
```

```python
# folium.Map.save() 를 이용해서 html 파일로 저장
    m.save('map.html')
    webbrowser.open('map.html') 

if __name__ == '__main__':
    main()
```
