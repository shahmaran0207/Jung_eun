```python
#강원도 범죄율 데이터 전처리
gangwon_cr=gangwon_cr.transpose()
gangwon_cr=gangwon_cr.drop(gangwon_cr.index[0])
gangwon_cr.columns=['강원도']

#경기도 범죄율 데이터 전처리
gunggi_cr=gunggi_cr.transpose()
gunggi_cr=gunggi_cr.drop(gunggi_cr.index[0])
gunggi_cr.columns=['경기도']

#경상남도 범죄율 데이터 전처리
gungnam_cr=gungnam_cr.transpose()
gungnam_cr=gungnam_cr.drop(gungnam_cr.index[0])
gungnam_cr.columns=['경상남도']

#경상북도 범죄율 데이터 전처리
gungbuk_cr=gungbuk_cr.transpose()
gungbuk_cr=gungbuk_cr.drop(gungbuk_cr.index[0])
gungbuk_cr.columns=['경상북도']

#광주광역시 범죄율 데이터 전처리
gwangju_cr=gwangju_cr.transpose()
gwangju_cr=gwangju_cr.drop(gwangju_cr.index[0])
gwangju_cr.columns=['광주광역시']

#대구광역시 범죄율 데이터 전처리
daegu_cr=daegu_cr.transpose()
daegu_cr=daegu_cr.drop(daegu_cr.index[0])
daegu_cr.columns=['대구광역시']

#대전광역시 범죄율 데이터 전처리
daejeon_cr=daejeon_cr.transpose()
daejeon_cr=daejeon_cr.drop(daejeon_cr.index[0])
daejeon_cr.columns=['대전광역시']

#부산광역시 범죄율 데이터 전처리
busan_cr=busan_cr.transpose()
busan_cr=busan_cr.drop(busan_cr.index[0])
busan_cr.columns=['부산광역시']

#울산광역시 범죄율 데이터 전처리
ulsan_cr=ulsan_cr.transpose()
ulsan_cr=ulsan_cr.drop(ulsan_cr.index[0])
ulsan_cr.columns=['울산광역시']

#서울특별시 범죄율 데이터 전처리
seoul_cr=seoul_cr.transpose()
seoul_cr=seoul_cr.drop(seoul_cr.index[0])
seoul_cr.columns=['서울특별시']

#인천광역시 범죄율 데이터 전처리
incheon_cr=incheon_cr.transpose()
incheon_cr=incheon_cr.drop(incheon_cr.index[0])
incheon_cr.columns=['인천광역시']

#전라남도 범죄율 데이터 전처리
jeonnam_cr=jeonnam_cr.transpose()
jeonnam_cr=jeonnam_cr.drop(jeonnam_cr.index[0])
jeonnam_cr.columns=['전라남도']

#전라북도 범죄율 데이터 전처리
jeonbuk_cr=jeonbuk_cr.transpose()
jeonbuk_cr=jeonbuk_cr.drop(jeonbuk_cr.index[0])
jeonbuk_cr.columns=['전라북도']

#제주특별자치도 범죄율 데이터 전처리
jeju_cr=jeju_cr.transpose()
jeju_cr=jeju_cr.drop(jeju_cr.index[0])
jeju_cr.columns=['제주특별자치도']

#충청남도 범죄율 데이터 전처리
chungnam_cr=chungnam_cr.transpose()
chungnam_cr=chungnam_cr.drop(chungnam_cr.index[0])
chungnam_cr.columns=['충청남도']

#충청북도 범죄율 데이터 전처리
chungbuk_cr=chungbuk_cr.transpose()
chungbuk_cr=chungbuk_cr.drop(chungbuk_cr.index[0])
chungbuk_cr.columns=['충청북도']
```
